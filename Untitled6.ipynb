### 1.自制数据集，数据增强

```python
import tensorflow as tf
from PIL import Image
import numpy as np
import os
```

```python
train_path = './dataset/MNIST_FC/mnist_image_label/mnist_train_jpg_60000/' # 训练集图片路径
train_txt = './dataset/MNIST_FC/mnist_image_label/mnist_train_jpg_60000.txt' # 训练集图片与标签文件
x_train_savepath = './dataset/MNIST_FC/mnist_image_label/mnist_x_train.npy' # 训练集特征存储文件
y_train_savepath = './dataset/MNIST_FC/mnist_image_label/mnist_y_train.npy' # 训练集标签存储文件

test_path = './dataset/MNIST_FC/mnist_image_label/mnist_test_jpg_10000/'
test_txt = './dataset/MNIST_FC/mnist_image_label/mnist_test_jpg_10000.txt'
x_test_savepath = './dataset/MNIST_FC/mnist_image_label/mnist_x_test.npy'
y_test_savepath = './dataset/MNIST_FC/mnist_image_label/mnist_y_test.npy'
```

```python
def generateds(path, txt): # 图片路径， 标签图片
    f = open(txt, 'r') # 只读形式打开文件
    contents = f.readlines() # 读取文件中所有行
    f.close() # 关闭txt文件
    x, y_ = [], [] # 建立空列表
    for content in contents: # 逐行取出
        value = content.split() # 以空格分开，图片路径为value[0],标签为value[1]，存入列表
        img_path = path + value[0]  # 拼出图片路径和文件名
        img = Image.open(img_path) # 读入图片
        img = np.array(img.convert('L')) # 图片变为8位宽灰度值的np.array格式
        img = img / 255. # 数据归一化（实现预处理）
        x.append(img) # 归一化后的数据，贴到列表x
        y_.append(value[1]) # 标签贴到列表y_
        print('loading:' + content) # 打印状态提示
        
    x = np.array(x) # 变为np.array格式
    y_ = np.array(y_) # 变为np.array格式
    y_ = y_.astype(np.int64) # 变为64位整型
    return x, y_
```

```python
if os.path.exists(x_train_savepath) and os.path.exists(y_train_savepath) and os.path.exists(x_test_savepath) and os.path.exists(y_test_savepath):
    print('-------------Load Datasets-----------------')
    x_train_save = np.load(x_train_savepath)
    y_train = np.load(y_train_savepath)
    x_test_save = np.load(x_test_savepath)
    y_test = np.load(y_test_savepath)
    x_train = np.reshape(x_train_save, (len(x_train_save), 28, 28))
    x_test = np.reshape(x_test_save, (len(x_test_save), 28, 28))
else:
    print('-------------Generate Datasets-----------------')
    x_train, y_train = generateds(train_path, train_txt)
    x_test, y_test = generateds(test_path, test_txt)
    
    print('-------------Save Datasets-----------------')
    x_train_save = np.reshape(x_train, (len(x_train), -1))
    x_test_save = np.reshape(x_test, (len(x_test), -1))
    np.save(x_train_savepath, x_train_save)
    np.save(y_train_savepath, y_train)
    np.save(x_test_savepath, x_test_save)
    np.save(y_test_savepath, y_test)
```

```python
# baseline
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['sparse_categorical_accuracy'])

model.fit(x_train, y_train, batch_size=32, epochs=5, validation_data=(x_test, y_test),
         validation_freq=1)

model.summary()
```

## 2.数据增强，扩充数据集

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
print('before:', x_train.shape)
```

```python
# 对图片进行变形
image_gen_train = ImageDataGenerator(
    rescale=1. / 1.,  # 如为图像，分母为255时，可归至0～1 （所有数据乘以该数值）
    rotation_range=45,  # 随机45度旋转 （随机旋转角度范围）
    width_shift_range=.15,  # 宽度偏移 （随机宽度偏移量）
    height_shift_range=.15,  # 高度偏移 （随机高度偏移量）
    horizontal_flip=False,  # 水平翻转 （是否随机水平翻转）
    zoom_range=0.5  # 将图像随机缩放阈量50％ （随机缩放的范围 【1-n, 1+n】）
)

# 样本数， 宽， 高， 通道数
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
print('增加一个维度:', x_train.shape)
image_gen_train.fit(x_train) # 这里的fit需要4维数据
```

```python
x_train.shape # 只是对图片做了变形操作， 以下的image_gen_train.flow 是否扩充了数量？
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['sparse_categorical_accuracy'])

model.fit(image_gen_train.flow(x_train, y_train, batch_size=32), epochs=5,
         validation_data=(x_test, y_test),
         validation_freq=1)

model.summary()
```

```python
from matplotlib import pyplot as plt
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['sparse_categorical_accuracy'])

# 断点
checkpoint_save_path = "./checkpoint/mnist.ckpt" # 模型文件
if os.path.exists(checkpoint_save_path + '.index'): # 模型索引表，由于在生成模型文件同时会同步生成索引表，故可以通过索引表来判断模型是否生成成功
    print('-------------load the model-----------------')
    # 读取模型参数
    model.load_weights(checkpoint_save_path)

# 回调函数
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_save_path, # 模型路径
                                                save_weights_only=True, # 是否只保留模型参数
                                                save_best_only=True) # 是否只保留最优结果

history = model.fit(x_train, y_train, batch_size=32, epochs=5, validation_data=(x_test, y_test),
                   validation_freq=1, callbacks=[cp_callback])

model.summary()

# 4.参数提取，将参数存入文本
np.set_printoptions(threshold=np.inf) # 直接打印会有很多..., 这里np.inf无限大 不省略信息
print(model.trainable_variables) # 返回模型中所有可训练参数

file = open('./weights.txt', 'w')
for v in model.trainable_variables:
    file.write(str(v.name) + '\n')
    file.write(str(v.shape) + '\n')
    file.write(str(v.numpy()) + '\n')
file.close()

# 显示训练集和验证集的acc和loss曲线
acc = history.history['sparse_categorical_accuracy']
val_acc = history.history['val_sparse_categorical_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.subplot(1, 2, 1) # 将图像划成一行两列，此处划出第一列
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy') # 图标题
plt.legend() # 图例

plt.subplot(1, 2, 2) # 此处划出第二列
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
model_save_path = './checkpoint/mnist.ckpt'

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.load_weights(model_save_path)

# 预测图片的数量
preNum = int(input("input the number of test pictures:"))

for i in range(preNum):
    image_path = input("the path of test picture:")
    img = Image.open(image_path)
    img = img.resize((28, 28), Image.ANTIALIAS) # # 输入尺寸任意，resize成标准尺寸28*28
    img_arr = np.array(img.convert('L')) # 灰度图像
    
    # 预处理
    # 方法1
    # img_arr = 255 - img_arr # 颜色取反（黑底白字与白底黑字）
    # 方法2 转换为高对比度图片 遍历每一个像素点 这种方式保留图片有用信息的同事，滤去了背景噪声，图片更干净
    for i in range(28):
        for j in range(28):
            if img_arr[i][j] < 200: # 阈值
                # 纯白色
                img_arr[i][j] = 255
            else:
                # 纯黑色
                img_arr[i][j] = 0
    
    img_arr = img_arr / 255.0 # 归一化
    x_predict = img_arr[tf.newaxis, ...] # 相当于reshape成(1， 28， 28)
    result = model.predict(x_predict) # 根据输入特征预测结果
    
    pred = tf.argmax(result, axis=1)
    
    print('\n')
    tf.print(pred)
    
    # input the number of test pictures: 10
    # the path of test picture: ./dataset/MNIST_FC/6.png
```
